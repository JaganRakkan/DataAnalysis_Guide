## Data preparation

```python
# Example of dropping columns
# df.drop(['Opening date'], axis=1)
df = df[['coaster_name','Location', 'Status','Manufacturer','year_introduced',
        'latitude', 'longitude','Type_Main','opening_date_clean','speed_mph',
        'height_ft','Inversions_clean', 'Gforce_clean']].copy()
```

```python
# New dataframe
df.head()
```

```python
df.shape # now dataset columns changed 56 to 13
```

```python
# changing data type of opjeact to date

df['opening_date_clean'] = pd.to_datetime(df.opening_date_clean)
```

```python
df.dtypes
```

```python
# Renaming our columns
df = df.rename(columns={
                'coaster_name':'Coaster_Name',
                'year_introduced':'Year_Introduced',
                'opening_date_clean':'Opening_Date',
                'speed_mph':'Speed_mph',
                'height_ft':'Height_ft',
                'Inversions_clean':'Inversions',
                'Gforce_clean':'Gforce'
})
```

```python
# checking Null values
df.isna().sum()
```

```python
# checking duplicates
df.loc[df.duplicated()]
```

```python
# Check for duplicate coaster name

df.loc[df.duplicated(subset=['Coaster_Name'])].head(10)
```

```python
# using query function to check how it is duplicated

df.query('Coaster_Name == "Crystal Beach Cyclone"')
```

* Only year_introduced value is differ from two rows other things remain same

```python
# So we are going to check the duplicates that contains same value

df.duplicated(subset=['Coaster_Name','Location','Opening_Date'])
```

```python
df.duplicated(subset=['Coaster_Name','Location','Opening_Date']).sum()
```

```python
# inverse of above cell

~df.duplicated(subset=['Coaster_Name','Location','Opening_Date'])
```

```python
df_new = df.loc[~df.duplicated(subset=['Coaster_Name','Location','Opening_Date'])].reset_index(drop=True)
```

```python
df_new.shape
```

## Feature understanding

```python
# It's show us how many unique values present in the data

df_new['Year_Introduced'].value_counts()
```

```python
# plotting the top years of coasters introduced

ax = df_new['Year_Introduced'].value_counts().head(10).plot(kind="bar",title="Top 10 years of coasters lanuch")
ax.set_xlabel('lanuch year')
ax.set_ylabel('count')
```

```python
# Distribution of coaster speed

speed_dis = df_new['Speed_mph'].plot(kind='hist',bins=20,title='coaster speed')
speed_dis.set_xlabel('lanuch year')
```

```python
speed_dis = df_new['Speed_mph'].plot(kind='kde',title='coaster speed')
speed_dis.set_xlabel('lanuch year')
```

#### Find what are locations has the fastest roller coaster in the data?

```python
df['Location'].value_counts()
```

```python
df_new.query('Location != "Other"')
```

```python
df_new.query('Location != "Other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean','count'])
```

```python
df_new.query('Location != "Other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean','count']) \
    .query('count >= 10') \
    .sort_values('mean')
```

```python
ax = df_new.query('Location != "Other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean','count']) \
    .query('count >= 10') \
    .sort_values('mean')['count'] \
    .plot(kind='barh', figsize=(12, 5), title='Average Coast Speed by Location')
ax.set_xlabel('Average Coaster Speed')
plt.show()
```

```python
ax = df_new.query('Location != "Other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean','count']) \
    .query('count >= 10') \
    .sort_values('mean')['mean'] \
    .plot(kind='barh', figsize=(12, 5), title='Average Coast Speed by Location')
ax.set_xlabel('Average Coaster Speed')
plt.show()
```
